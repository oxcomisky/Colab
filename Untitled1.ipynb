```python
class sub:
    vertical = 0
    horizontal = 0
    aim = 0
    

def move(sub, command):
    instr = command.split(" ")
    if(instr[0]=="forward"):
        sub.horizontal += int(instr[1])
        sub.vertical += (int(instr[1]) * sub.aim)
    elif(instr[0]=="reverse"):
        sub.horizontal -= int(instr[1])
    elif(instr[0]=="up"):
        sub.aim -= int(instr[1])
        
    elif(instr[0]=="down"):
        sub.aim += int(instr[1])
        

sub = sub()
#parse file and collect commands, then remove newlines
f = open("/content/drive/MyDrive/Day2CodingChallenge", mode='r')
list1 = []
i = 0
while i < 1000:
    list1.append(f.readline())
    i += 1
commands = []
for command in list1:
    commands.append(command.replace("\n", ""))
print(commands)

for command in commands:
     move(sub, command)
     print("command: " + command)
     print("vertical: " + str(sub.vertical))
     print("horizontal: " + str(sub.horizontal))
     print("aim: " + str(sub.aim))
     print("position: " + str(sub.vertical * sub.horizontal))

```
